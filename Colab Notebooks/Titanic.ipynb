```
import pandas as pd
import matplotlib.pyplot as plt

#Kaggleのタイタニック生存者問題

#Google Driveからデータを読み込む
from google.colab import drive
drive.mount('/content/gdrive')
df = pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/data/train.csv').replace("male",0).replace("female",1)

df["Age"].fillna(df.Age.median(), inplace=True)
df["Embarked"].fillna("S", inplace=True)
df.Embarked = df.Embarked.replace("S", 0)
df.Embarked = df.Embarked.replace("C", 1)
df.Embarked = df.Embarked.replace("Q", 2)

print(df.isnull().sum())
print(df.describe())

#生き残った人のグラフを見てみる
#緑が生存、青が死亡
#split_data = []
#for survived in [0,1]:
#    split_data.append(df[df.Survived==survived])


#客室ごとのグラフ
#左から1等、2等、3等
#temp = [i["Pclass"].dropna() for i in split_data]
#plt.hist(temp, histtype="barstacked", bins=3)

#年齢ごと
#temp = [i["Age"].dropna() for i in split_data]
#plt.hist(temp, histtype="barstacked", bins=16)

#学習データ
train_data = df.drop(["Survived","Name", "SibSp", "Parch", "Ticket", "Cabin"], axis=1) # Pclass以降の変数
xs = train_data.values
y  = df["Survived"].values  # 正解データ

#ランダムフォレスト
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators = 100)

# 学習
forest = forest.fit(xs, y)

test_df= pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/data/test.csv').replace("male",0).replace("female",1)
# 欠損値の補完
test_df["Age"].fillna(df.Age.median(), inplace=True)
test_df.Embarked = test_df.Embarked.replace("S", 0)
test_df.Embarked = test_df.Embarked.replace("C", 1)
test_df.Embarked = test_df.Embarked.replace("Q", 2)
test_df["Fare"].fillna(df.Fare.median(), inplace=True)

test_df2 = test_df.drop(["Name", "SibSp", "Parch", "Ticket", "Cabin"], axis=1)

#print(test_df.isnull().sum())
#print(test_df.describe())

#予測の出力
#test_data = test_df2.values
xs_test = test_df2.values
output = forest.predict(xs_test)

#csvファイルにする
import csv
with open("/content/gdrive/My Drive/Colab Notebooks/data/predict_result_data.csv", "w") as f:
    writer = csv.writer(f, lineterminator='\n')
    writer.writerow(["PassengerId", "Survived"])
    for pid, survived in zip(test_data[:,0].astype(int), output.astype(int)):
        writer.writerow([pid, survived])
```
